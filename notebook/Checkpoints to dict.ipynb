```python
import numpy as np
from tensorpack.tfutils.common import get_op_tensor_name
import six
from tensorpack.tfutils import varmanip
import tensorflow as tf
import tensorflow._api.v2.compat.v1 as tf
import os
```

```python
# https://tensorpack.readthedocs.io/en/latest/tutorial/save-load.html
# Help here: https://github.com/tensorpack/tensorpack/blob/master/scripts/dump-model-params.py
# https://github.com/tensorpack/tensorpack/blob/master/scripts/ls-checkpoint.py

### COMA AUTOENCODER LAT 32
basedir = "/home/oole/conv-track/final_model_dir/coma-autoencoder-lat32/checkpoints"
dict_dir = basedir.replace("checkpoints", "dicts")

names = ["tp_coma_autoencoder_lat-32_cheb_edge-loss_220209-1432",
        "tp_coma_autoencoder_lat-32_cheb_l1-only-220209-1247",
        "tp_coma_autoencoder_lat-32_spiral_edge-loss_220209-1314",
        "tp_coma_autoencoder_lat-32_spiral_l1-only-220209-1252"]

epoch_300_step = "model-238080"

for name in names:
    checkpoint_path = varmanip.get_checkpoint_path(os.path.join(basedir, name, epoch_300_step))
    print(checkpoint_path)
    checkpoint_vars_dict =varmanip.load_checkpoint_vars(checkpoint_path)
    mesh2feature_var_dict =  {"mesh2feature/"+get_op_tensor_name(k)[1]: v for k,v in six.iteritems(checkpoint_vars_dict) if "tower0" not in get_op_tensor_name(k)[1] and "global_step" not in get_op_tensor_name(k)[1]}
    varmanip.save_chkpt_vars(mesh2feature_var_dict, os.path.join(dict_dir, name + "-300epochs.npz"))
# reader = tf.train.NewCheckpointReader(checkpoint_path)
# var_dict = reader.get_variable_to_shape_map()
```
